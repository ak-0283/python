### Advance operation with business examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sales_data = np.array([
    [1, 1280349,23421,4544344,34912312],
    [2, 1228034,23221,4941344,34512312],
    [3, 1280234,24221,4543344,34412312],
    [4, 1280364,23221,4503344,34112312],
    [5, 1428034,23021,4543044,34012312],
])


print("=== zomato sales analysis === ")
print("\n sales data shape",sales_data.shape)
print("\n sample data for 1st 3 restu.",sales_data[:3])
print("\n sample data for 1st 3 restu.",sales_data[:,1:])
```

### Total sales Per Year

```python
print(np.sum(sales_data,axis=0))
yearly_total=np.sum(sales_data[:,1:],axis=0)
print(yearly_total)

### minimum sales per restuarant
min_sales=np.min(sales_data[:,1:],axis=1)
print(min_sales)


### maximum sales per year
max_sales=np.max(sales_data[:,1:],axis=0)
print(max_sales)


### Average Sales Per Restuarant
avg_sales=np.mean(sales_data[:,1:],axis=1)
print(avg_sales)
```

### Cummulative Sales

```python
cumsum =  np.cumsum(sales_data[:,1:],axis=1)
print(cumsum)


```

```python
plt.figure(figsize=(8,6))
plt.plot(np.mean(cumsum,axis=0))
plt.title("Average cummulative sales all restuarant")
plt.xlabel("years")
plt.ylabel("sales")
plt.grid(True)
plt.show()
```
